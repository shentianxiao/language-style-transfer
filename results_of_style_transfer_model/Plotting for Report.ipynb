```python
import pandas as pd
import seaborn as sns
import matplotlib as plt
```

```python
res = pd.read_csv("/mnt/c/Users/kikibox/Documents/NLP/language-style-transfer/author_identification_code/"+"shakespeare_hemingway_f1_score_overtime.csv")
res
```

```python
plt.rcParams["figure.figsize"] = (10,10)
ax = sns.lineplot(x="epoch", y="f1_score_weighted", hue="prediction_type",
                   estimator=None, lw=1,
                  data=res).set_title('F1 Score (weighted) for Author Identification \n of Style Transfer Model Results and baselines \n across all Epochs (Shakespeare-Hemingway Pairing)', fontsize = 15)
```

```python
ax = sns.lineplot(x="epoch", y="precision_score_weighted", hue="prediction_type",
                   estimator=None, lw=1,
                  data=res)
```
